```python
my_dict = {"a": 1, "b": 2, "c": 3}
print(my_dict.keys())        # Get all keys
print(my_dict.values())      # Get all values
print(my_dict.items())       # Get key-value pairs
print(my_dict.get("b"))      # Get value of key 'b'
my_dict.update({"d": 4})     # Add new key-value
```

```python
my_list = [10, 20, 30, 40]
my_list.append(50)           # Add element
my_list.remove(20)           # Remove element
print(my_list.index(30))     # Find index
print(my_list.count(40))     # Count occurrences
my_list.sort()               # Sort list
```

```python
my_tuple = (1, 2, 3, 4, 2)
print(my_tuple.count(2))     # Count occurrences
print(my_tuple.index(3))     # Find index
print(len(my_tuple))         # Length
print(max(my_tuple))         # Max value
print(min(my_tuple))         # Min value
```

```python
my_set = {1, 2, 3}
my_set.add(4)                # Add element
my_set.remove(2)             # Remove element
print(my_set.union({5, 6}))  # Union
print(my_set.intersection({3, 4})) # Intersection
print(my_set.difference({1, 4}))   # Difference
```

```python
x = 10

# If
if x > 5:
    print("x is greater than 5")

# If-Else
if x % 2 == 0:
    print("Even")
else:
    print("Odd")

# If-Elif-Else
if x < 0:
    print("Negative")
elif x == 0:
    print("Zero")
else:
    print("Positive")

# Nested If-Else
if x > 0:
    if x < 20:
        print("x is between 1 and 19")
    else:
        print("x is 20 or more")
else:
    print("x is not positive")
```

```python
# Break
for i in range(5):
    if i == 3:
        break
    print(i)

# Continue
for i in range(5):
    if i == 3:
        continue
    print(i)

# Pass
for i in range(5):
    if i == 3:
        pass  # Placeholder
    print(i)
```
